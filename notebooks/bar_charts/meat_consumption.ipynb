# Bar chart showing the meat consumption per capita (top 10 countries)

## Get the data

Let's load the data, filter by year (2023), sort by meat consumption value in descending order and select the first ten records.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/bianconif/data_for_visualisation/refs/heads/master/top-10-meat-consumption-countries-2023.csv',comment='#')
df = df[df['Year'] == 2023]
df.sort_values(by='Value', ascending=False, inplace=True)
df = df.iloc[0:10]
print(df)
```

Now we create an empty figure (`fig`) with one sub-plot (`ax`) using [Matplotlib's object oriented interface](https://matplotlib.org/stable/users/explain/figure/api_interfaces.html) [1]. We then use the subplot as the canvas where to generate the bar chart using Matplotlib's [`matplotlib.axes.Axes.bar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html) method. Finally we rotate the x-axis tick labels vertically to avoid them from overlapping. Note that we keep the value returned by [`Axes.bar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html) (`bars`) for future use.

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
bars = ax.bar(x=df['Country'], height=df['Value'])
ax.tick_params('x', rotation=90)
```

## Changing the bar's colour

Our first tweak consists of changing the bar's edge and fill colour. We pick the fill colour from one of Matplotlib's predefined palettes - the Tableau palette in this case. Note that we modify the colours a posteriori by leveraging on the `set_color` and `set_edgecolor` methods. Alternatively, we could have set the colours a priori by passing them as arguments to Matplotlib's [matplotlib.axes.Axes.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html) (parameters `facecolor` and `edgecolor`). 

```python
fill_colour = 'tab:orange'
edge_colour = 'black'
for bar in bars:
    bar.set_color(fill_colour)
    bar.set_edgecolor(edge_colour)
display(fig)
```
